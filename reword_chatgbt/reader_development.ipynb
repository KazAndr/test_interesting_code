```python
import pandas as pd
import sqlite3
```

```python
# Замените 'your_database.db' на путь к вашему файлу базы данных SQLite
database_path = 'reword_en.backup'
```

```python
# Подключение к базе данных
conn = sqlite3.connect(database_path)
```

```python
# Создание объекта cursor, который позволяет вам выполнять SQL команды
cursor = conn.cursor()
```

```python
# SQL запрос для выбора всех таблиц в базе данных
query = "SELECT name FROM sqlite_master WHERE type='table'"
```

```python
cursor.execute(query)
tables = cursor.fetchall()

# Словарь для хранения структуры базы данных
db_structure = {}

for table_name in tables:
    table_name = table_name[0]  # Получаем имя таблицы
    # Получаем список столбцов для каждой таблицы
    cursor.execute(f"PRAGMA table_info({table_name});")
    columns = cursor.fetchall()
    # Добавляем в словарь, где ключ это имя таблицы, а значение — список имен столбцов
    db_structure[table_name] = df = pd.read_sql_query(f"SELECT * FROM {table_name}", conn)
```

```python
# Закрытие соединения с базой данных
conn.close()
```
