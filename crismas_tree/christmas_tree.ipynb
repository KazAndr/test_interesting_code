```python
import pandas as pd
import numpy as np
import os
import re
import glob
import matplotlib.pyplot as plt
```

```python
def read_header(filename):
    with open(filename, 'r') as file:
        header = {}
        for i in range(20):
            try:
                a, b = file.readline().split()
                header[a] = b
            except ValueError:
                continue
    return header
```

```python
file = sorted(glob.glob('*.csv'))
```

```python
header = read_header(file[0])
table_pulse = pd.read_csv(file[0], sep = '\t', header = 10, names = range(int(header['numpuls'])))
```

```python
dic_pulse = {}
for inx in range(len(table_pulse.T)):
    dic_pulse[inx] = max(table_pulse[inx])
```

```python
import operator
sorted_dic = sorted(dic_pulse.items(), key=operator.itemgetter(1))
```

```python
list_index = []
for tup in reversed(sorted_dic):
    list_index.append(tup[0])
```

```python
table_pulse[66][100:200].plot()
plt.show()
```

```python
christmas_tree = []
start = 120
stop = 180
for inx, val in enumerate(list_index):
    christmas_tree.extend(table_pulse[val][start:stop])
```

```python
plt.close()
plt.plot(christmas_tree[50:5000])
plt.plot((-1)*christmas_tree[50:5000])

plt.show()
```

```python
start =100
stop = 5000
christmas_tree = np.asarray(christmas_tree)
x = range(len(christmas_tree[start:stop]))
y1 = christmas_tree[start:stop:]
y2 = -1*christmas_tree[start:stop:]
```

```python
plt.close()
plt.style.use('dark_background')
fig, ax1 = plt.subplots(1, 1, sharex=True)
plt.plot(y1, x, color = 'g')
plt.plot(y2, x, color = 'g')
plt.xlim(-70, 70)
#ax1.fill_between(x, y2, y1, facecolor='green', interpolate=False)
plt.savefig('christmas_tree.png', format='png', dpi=1000)
plt.show()
```
